```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm, probplot

plt.figure(figsize=(8,5))
sns.histplot(data['Temp_C'], kde=True, stat="density", bins=30, color='skyblue')
mean = data['Temp_C'].mean()
std = data['Temp_C'].std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'r', linewidth=2)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(6,6))
probplot(data['Temp_C'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Temperature')
plt.show()
```

Ploting Graphs

```python
#Temperature
plt.figure(figsize=(8,5))
plt.hist(data['Temp_C'], bins=30, color='skyblue', edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#Scatter plot: Temperature vs Humidity
plt.figure(figsize=(8,5))
plt.scatter(data['Temp_C'], data['Rel Hum_%'], alpha=0.5, color='black')
plt.title('Temperature vs Relative Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.grid()
plt.show()
```

```python
#Line plot: Temperature over time 
if 'Date/Time' in data.columns:
    data['Date/Time'] = pd.to_datetime(data['Date/Time'])
    plt.figure(figsize=(12,6))
    plt.plot(data['Date/Time'], data['Temp_C'], color='purple')
    plt.title('Temperature Over Time')
    plt.xlabel('Date/Time')
    plt.ylabel('Temperature (C)')
    plt.grid()
    plt.show()

```

```python
#Bar plot: Weather Condition counts
plt.figure(figsize=(10,5))
counts = data['Weather Condition'].value_counts()
plt.bar(counts.index, counts.values, color='orange')
plt.xticks(rotation=90)
plt.title('Weather Condition Counts')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.grid()
plt.show()
```
